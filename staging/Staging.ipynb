```python
import urllib.request

url = 'https://static.openfoodfacts.org/data/en.openfoodfacts.org.products.csv.gz'
volume_path = "/Volumes/openfoodfactslakehouse/staging/dataset_csv/products_raw.csv.gz"

try:
    urllib.request.urlretrieve(url, volume_path)
    print(f"File successfully downloaded to {volume_path}")
except Exception as e:
    print(f"An error occurred: {e}")
```

```python
df = spark.read \
    .option("header", True) \
    .option("delimiter", "\t") \
    .option("multiLine", True) \
    .option("escape", "\"") \
    .option("compression", "gzip") \
    .csv("dbfs:/Volumes/openfoodfactslakehouse/staging/dataset_csv/products_raw.csv.gz")

(
    df.write
    .format("delta")
    .mode("overwrite")
    .option("mergeSchema", "true")
    .saveAsTable("openfoodfactslakehouse.staging.products_table")
)
```
